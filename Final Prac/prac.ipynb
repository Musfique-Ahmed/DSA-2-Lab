```python
from heapq import heappop, heappush


def prims_algo(adj_list, src):

    x = len(adj_list)
    
    key = [float('inf')] * x
    parent = [-1] * x
    treevertex = [False] * x

    key[src] = 0

    meanheap = [(key[src], src)]

    while meanheap:
        k, u = heappop(meanheap)

        if treevertex[u]: continue

        treevertex[u] = True

        for v,w in adj_list[u]:
            if not treevertex[v] and key[v] > w:
                key[v] = w
                parent[v] = u
                heappush(meanheap, (key[v], v))

    return key, parent



with open("C:\Musfique's Folder\Python\DSA-2-Lab\Final Prac\sparseGraph.txt", "r") as f:
    # print(f.readline)
    n,m = map(int, f.readline().split())
    # print(n,m)

    adj_list = {i:[] for i in range(n)}
    # print(adj_list)

    for _ in range(m):
        u, v, w = map(int, f.readline().split())

        adj_list[u].append((v,w))
        adj_list[v].append((u,w))

    key, parent = prims_algo(adj_list, 0)

    print(f"The sum of MST is: {sum(key)}")

    print(adj_list)
```

```python
with open("C:\Musfique's Folder\Python\DSA-2-Lab\Final Prac\sparseGraph.txt", "r") as f:
    n,m = map(int, f.readline().split())
    edge_lst = []
    for _ in range(m):
        u,v,w = map(int, f.readline().split())
        edge_lst.append((w,u,v))


parent = [i for i in range(n)] #[0,1,2,3,4,5,...,n-1]
rank  = [0] * n #[0,0,0,0,0,....]

def find(u):
    if parent[u] != u:
        parent[u] = find(parent[u])
    return parent[u]

def union(x, y):
    xroot = find(x)
    yroot = find(y)

    if rank[xroot] < rank[yroot]:
        parent[xroot] = yroot
    elif rank[xroot] > rank[yroot]:
        parent[yroot] = xroot
    else:
        parent[xroot] = yroot
        rank[yroot] += 1


def kruskal_algo(edge_list):
    # step-1: sort kora
    edge_list.sort()
    # step-2: initial jinish porto banano
    mst_edge = []
    total_weight = 0

    # step-3: parent and rank array func er baire banano
    #step - 4: find and union func banano

    # step-5: loop in the edge_list and connect the edges -> compair parents, union, total weight increase, add to most_edge list
    for w,u,v in edge_list:
        if find(u) != find(v):
            union(u,v)
            total_weight += w
            mst_edge.append((u,v,w))

    return total_weight, mst_edge

w, e = kruskal_algo(edge_lst)

print(w)
print(e)
```
